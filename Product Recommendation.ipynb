```python
import pywhatkit as kit
from datetime import datetime
import pandas as pd


df = pd.read_excel('Customer Purchase.xlsx')
products = pd.read_excel('products.xlsx')

def extract_items(items_str):
    items = items_str.split(',')
    item_list = []
    for item in items:
        item_name = item.split('(')[0].strip() 
        item_list.append(item_name)
    return item_list


def recommended_product(items):
    purchased_categories = products[products['product_name'].isin(items)]['category'].unique()
    recommended_products = products[products['category'].isin(purchased_categories)]['product_name'].tolist()
    return recommended_products

customer_purchase_df = df.copy()
customer_purchase_df['Item_List'] = customer_purchase_df['Items'].apply(extract_items)


product_customers = {}
for index, row in customer_purchase_df.iterrows():
    for item in row['Item_List']:
        if item not in product_customers:
            product_customers[item] = []
        customer_details = {
            'Customer ID': row['Customer ID'],
            'Name': row['Name'],
            'Contact': row['Contact']
        }
        product_customers[item].append(customer_details)


def get_customers_by_product(product_name):
    return product_customers.get(product_name, [])


mango_customers = get_customers_by_product('Mango')

for customer in mango_customers:
    contact_number = customer['Contact']
    formatted_number = f'+{contact_number}'  

    items_list = customer_purchase_df[customer_purchase_df['Customer ID'] == customer['Customer ID']]['Item_List'].values[0]
    
    recommended_items = ', '.join(recommended_product(items_list))

    message_body = f"Hello {customer['Name']},\n\nWe recommend the following products for you: {recommended_items}.\n\nBest Regards, Your Company."
    

    now = datetime.now()
    kit.sendwhatmsg(formatted_number, message_body, now.hour, now.minute + 1)
```
